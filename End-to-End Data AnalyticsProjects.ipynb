Read unzipped csv files and handle null values 

```python
import pandas as pd
df=pd.read_csv('orders.csv',na_values=['Not Available', 'unknown','nan'])
df
```

Convert dataframe column to lowercase and replace with underscore

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
```

Calculate new columns (Discount amount,Sale Price & Profit) in Dataframe

```python
df['discount_amount'] =df['list_price']* df['discount_percent']/100
df['sale_price']=df['list_price']-df['discount_amount']
df['profit']=df['sale_price']-df['cost_price']
df
```

```python
df.dtypes
```

Convert datatype of order_date column(object) into date

```python
df['order_date']=pd.to_datetime(df['order_date'],format='%Y-%m-%d')
```

```python
df.dtypes
```

Drop columns cost price,list price,discount percent

```python
df.drop(['cost_price','list_price','discount_percent'],axis=1,inplace=True) 
```
